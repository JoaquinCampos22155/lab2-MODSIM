## Teoría

1. Defina un retraso de primer orden y explique sus dos formulaciones

Es el tipo de comportamiento que muestra cómo una variable reacciona a un cambio en otra. No cambia de inmediato, sino que se ajusta poco a poco hacia un nuevo valor.Se representa como: Seguimiento de flujo, Búsqueda de objetivo.

2. Seguimiento de flujo (salida = stock/retardo)

Esta fórmula se usa cuando se modela algo y se sale de un sistema de forma gradual. Si se tiene stock, la salida será una fracción de esa cantidad, dependiendo del retardo. Por ejemplo, si tenemos 100 unidades y el retardo es de 10 días, salen 10 unidades por día.

3. Búsqueda de objetivo (flujo = (objetivo - stock)/retardo)

En este caso, el sistema intenta ajustar el stock hacia un valor deseado. Cuanto más lejos esté el stock del objetivo, más rápido se moverá. El tiempo de retardo define qué tan rápido ocurre ese ajuste. Si el stock es 70, el objetivo es 100 y el retardo es 10, el flujo será (100 - 70)/10 = 3 unidades por día hacia arriba.

4. Describa la condición de equilibrio (entrada = salida) y cómo los stocks se ajustan exponencialmente.

Un sistema está en equilibrio cuando lo que entra es igual a lo que sale. En ese punto, el stock no cambia.

por ejemplo, cuando entra más de lo que sale, el stock empieza a ajustarse poco a poco. Este ajuste sigue una curva exponencial.

5. Explique por qué los sistemas lineales permiten la superposición (las soluciones se pueden sumar).

En un sistema lineal, los efectos de varias entradas se pueden analizar por separado y luego sumar. Es como si cada entrada tuviera su propio efecto, y el resultado final es la suma de todos. Esto hace que estos sistemas sean más fáciles de entender y modelar.

6. Dado un stock con un valor inicial de 100, una entrada de 5/día y un retraso de 10 días:

a. Calcule el valor de equilibrio del stock.

Para que el sistema llegue a equilibrio, la salida debe igualar la entrada. Usando seguimiento de flujo (salida = stock / retardo), se iguala:


$entrada=salida⇒5=stock/10⇒stock=50$


b. Dibuje la trayectoria del stock de salida a lo largo del tiempo.

Al inicio el stock es 100, pero como la entrada es de 5/día y el sistema tiende a bajar hacia 50 (el nuevo equilibrio), el stock disminuye gradualmente.
La trayectoria será una curva descendente suave que baja rápido



c. Compare los retrasos sin memoria y los retrasos por etapas (distribuciones de Erlang).

Retrasos sin memoria:
No importa cuánto tiempo ha pasado, la probabilidad de que algo salga sigue siendo la misma.

Retrasos por etapas:
Dividen el proceso en varias etapas pequeñas y ordenadas.


¿Cuándo usar cuál?

Usa retraso sin memoria cuando el tiempo de salida es impredecible

Usa retraso por etapas cuando hay un proceso con pasos fijos.

```python
# DIAGRAMA INCISO 6.b
import numpy as np
import matplotlib.pyplot as plt

entrada_diaria = 5
retardo_dias = 10
stock_inicial = 100
dias_simulacion = 60

tiempo = np.arange(0, dias_simulacion + 1)
stock = np.zeros(len(tiempo))
salida = np.zeros(len(tiempo))

stock[0] = stock_inicial

for t in range(1, len(tiempo)):
    salida[t-1] = stock[t-1] / retardo_dias
    stock[t] = stock[t-1] + entrada_diaria - salida[t-1]

salida[-1] = stock[-1] / retardo_dias

plt.figure(figsize=(10, 6))
plt.plot(tiempo, stock, label="Stock", linewidth=2)
plt.plot(tiempo, salida, label="Salida", linestyle="--", linewidth=2)
plt.axhline(y=50, color='gray', linestyle=':', label="Equilibrio esperado (Stock = 50)")
plt.title("Trayectoria del Stock y la Salida a lo largo del Tiempo")
plt.xlabel("Días")
plt.ylabel("Unidades")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Práctica

```python
import numpy as np
import matplotlib.pyplot as plt


# Stock inicial = 0 unidades
# Entrada constante = 10 unidades/día
# Salida = Stock / Retardo (retardo = 5 días)
# Paso de tiempo (Δt) = 1 día
# Tiempo total = 50 días


# Parámetros
dt = 1  # Paso de tiempo
dias = 50
tiempo = np.arange(0, dias + 1, dt)

# Condiciones del sistema
entrada_base = 10  # unidades/día
retardo = 5  # días
stock = [0]  # valor inicial
salida = []



for t in tiempo[:-1]:
    entrada = entrada_base
    salida_actual = stock[-1] / retardo
    nuevo_stock = stock[-1] + dt * (entrada - salida_actual)

    salida.append(salida_actual)
    stock.append(nuevo_stock)

equilibrio = entrada_base * retardo

# Gráfica
plt.figure(figsize=(10, 5))
plt.plot(tiempo, stock, label='Stock', color='green')
plt.plot(tiempo[:-1], salida, label='Salida', color='orange')
plt.axhline(equilibrio, linestyle='--', color='gray', label='Equilibrio esperado')

plt.xlabel('Tiempo (días)')
plt.ylabel('Unidades')
plt.title('Sistema con retardo de primer orden (Entrada constante)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Reiniciar variables
stock = [0]
salida = []

for t in tiempo[:-1]:
    entrada = 10 if t < 25 else 20
    salida_actual = stock[-1] / retardo
    nuevo_stock = stock[-1] + dt * (entrada - salida_actual)

    salida.append(salida_actual)
    stock.append(nuevo_stock)

eq_1 = 10 * retardo
eq_2 = 20 * retardo

# Gráfica
plt.figure(figsize=(10, 5))
plt.plot(tiempo, stock, label='Stock', color='green')
plt.plot(tiempo[:-1], salida, label='Salida', color='orange')
plt.axhline(eq_1, linestyle='--', color='gray', label='Equilibrio entrada=10')
plt.axhline(eq_2, linestyle='--', color='purple', label='Equilibrio entrada=20')
plt.axvline(25, linestyle=':', color='black', label='Cambio de entrada')

plt.xlabel('Tiempo (días)')
plt.ylabel('Unidades')
plt.title('Sistema con retardo de primer orden (Entrada cambia en día 25)')
plt.legend()
plt.grid(True)
plt.show()
```
